## Questions


I have chosen to investigate this specific question: **Can a player’s experience level and age predict how much time they spend playing on the Minecraft server?** I have chosen to investigate this specific question: Can a player’s experience level and age predict how much time they spend playing Minecraft in the dataset? It is likely that age and experience could play a factor in interest, or available free time, and would affect the amount of played hours on the server. Because `players.csv` already has the time spent on the server converted to a numeric value, we can disregard `sessions.csv`, and focus on the variables `experience`, `Age`, and `played_hours`.

We can begin by first taking a look at our variables of investigation. Since the second csv file, `sessions.csv` has the start and end times of playing, while the first csv file already has the data as played hours, we can disregard `sessions.csv` for our investigation and focus on `players.csv`. We will focus on the variables `experience`, `Age`, and `played_hours`. We can begin by selecting those variables and wrangling them into a tidy format:


```r
tidy_data <- minecraft_players_data |>
            select(experience, Age, played_hours) |>
            filter(Age != "NA", played_hours != "0" ) |>
            mutate(experience = factor(experience, 
                             levels = c("Beginner", "Amateur", "Regular", "Veteran", "Pro")))
tidy_data
```

```r
age_hours_plot <- tidy_data |>
            ggplot(aes(x = Age, y = played_hours)) +
            geom_point(alpha = 0.4) +
            labs(x = "Age of Players (Years)", y = "Hours Played by Players", title = "Age vs. Played Hours in Minecraft") +
            theme(text = element_text(size = 12))
age_hours_plot
```

```r
experience_hours_plot <- tidy_data |>
            ggplot(aes(x = experience, y = played_hours, fill = experience)) +
            geom_bar(stat = "identity") +
            labs(x = "Experience Level", y = "Hours Played by Players", title = "Experience Level vs. Played Hours in Minecraft") +
            theme(text = element_text(size = 12)) +
            scale_color_brewer(palette = "Set2")
experience_hours_plot
```
